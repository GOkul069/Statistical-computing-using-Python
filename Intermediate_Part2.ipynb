3. It appears that the data are organized with observations stored as columns rather than as rows. You can fix that. First, though, you can address the missing data. All of the NA values are stored in the first row. This row really belongs in a different data frame; it is a quarterly average of weekday MBTA ridership. 4 In this data, variables are stored in rows instead of columns. The different modes of transportation (commuter rail, bus, subway, boat, ...) are variables, providing information about each month’s average ridership. The months themselves are observations (Currently, months are listed as variable names; rather, they should be in their own column). You can tell which is which because as you go through time, the month changes, but the modes of transport offered by the T do not. As is customary, you want to represent variables in columns rather than rows.

Pivot the rows and columns of the mbta data so that all columns are variables of the data. This should result in 3 columns - mode, date, and number of riders in thousands ( thou_riders ).
What are the new dimensions of this data?

```python
if 'mode' in df_clean.columns:
    df_long = df_clean.melt(id_vars='mode', var_name='date', value_name='thou_riders')
else:
    df_long = df_clean.reset_index().rename(columns={'index': 'mode'}).melt(id_vars='mode', var_name='date', value_name='thou_riders')

df_long['thou_riders'] = pd.to_numeric(df_long['thou_riders'], errors='coerce')

try:
    df_long['date'] = pd.to_datetime(df_long['date'], format='%Y-%m')
except Exception:
    df_long['date'] = pd.to_datetime(df_long['date'], errors='coerce')

print("Reshaped dimensions:", df_long.shape)   
df_long.head()
```

### For this module we will be using the completejourney_py data sets that you saw in the lesson 4b reading:

1 Using the transactions and demographics data, how many of the 1,469,307 transactions do we have demographic information for?

```python
import pandas as pd
from completejourney_py import get_data
```

```python
cj = get_data()
print("Available cj_data tables:", list(cj.keys()))
```

```python
transactions = cj['transactions']
demographics = cj['demographics']
products = cj['products']
```

```python
merged_demo = transactions.merge(demographics, how='outer', indicator=True)
```

```python
common_col = 'household_id'  # typical name; change if different
num_with_demo = transactions.merge(demographics, how='inner', on=common_col).shape[0]
print("Transactions with demographic info:", num_with_demo, "out of", transactions.shape[0])
```

2 Using the transactions and demographics data, compute the total sales_value by age category to identify which age group generates the most sales.

```python
merge_td = transactions.merge(demographics, how='left', on=common_col)

age_cols = [c for c in demographics.columns if 'age' in c.lower()]
print("Possible age columns in demographics:", age_cols)

age_col = 'age' if 'age' in demographics.columns else (age_cols[0] if age_cols else None)
if age_col is None:
    raise RuntimeError("No age column found in demographics. Inspect demographics.columns.")
sales_by_age = merge_td.groupby(age_col, as_index=False)['sales_value'].sum().sort_values('sales_value', ascending=False)
print("Total sales by age group:\n", sales_by_age.head())
```

3 Identify all different products that contain “pizza” in their product_type description. Which of these products produces the greatest amount of total sales (compute total sales by product ID and product type)?

```python
pizza_mask = products['product_type'].str.contains('pizza', case=False, na=False)
pizza_products = products[pizza_mask].copy()
print("Number of product types containing 'pizza':", pizza_products.shape[0])

pizza_sales = (
    transactions
    .merge(pizza_products[['product_id', 'product_type']], how='inner', on='product_id')
    .groupby(['product_id', 'product_type'], as_index=False)
    .agg(total_sales=('sales_value', 'sum'))
    .sort_values('total_sales', ascending=False)
)
print("Top pizza products by total sales:\n", pizza_sales.head())
```

4 Identify all products that are categorized ( product_category ) as “pizza” but are considered a “snack” or “appetizer” (via product_type ). Which of these products ( product_id ) have the most number of sales (measured by quantity )?

```python
pizza_cat = products['product_category'].str.contains('pizza', case=False, na=False)
snack_app = products['product_type'].str.contains('snack|appetizer', case=False, na=False)
pizza_snack = products[pizza_cat & snack_app].copy()
print("Products categorized as pizza but product_type snack/appetizer:", pizza_snack.shape[0])

pizza_snack_sales = (
    transactions
    .merge(pizza_snack[['product_id', 'product_type']], how='inner', on='product_id')
    .groupby(['product_id', 'product_type'], as_index=False)
    .agg(total_quantity=('quantity', 'sum'))
    .sort_values('total_quantity', ascending=False)
)
print("Top pizza-as-snack/appetizer products by quantity sold:\n", pizza_snack_sales.head())
```

5 Identify all products that contain “peanut butter” in their product_type. How many unique products does this result in? For these products, compute the total sales_value by month based on the transaction_timestamp . Which month produces the most sales value for these products?

```python
pb_mask = products['product_type'].str.contains('peanut butter', case=False, na=False)
pb_products = products[pb_mask].copy()
num_unique_pb = pb_products['product_id'].nunique()
print("Unique 'peanut butter' products:", num_unique_pb)

tx_pb = transactions.merge(pb_products[['product_id','product_type']], how='inner', on='product_id')

tx_pb['transaction_timestamp'] = pd.to_datetime(tx_pb['transaction_timestamp'])
tx_pb['month'] = tx_pb['transaction_timestamp'].dt.strftime('%Y-%m')  # year-month
sales_by_month_pb = tx_pb.groupby('month', as_index=False)['sales_value'].sum().sort_values('sales_value', ascending=False)
print("Peanut butter sales by month (top months):\n", sales_by_month_pb)
```
