# Manipulating data

#### 1 Are there any missing values in this data? If so, which columns? For these columns, fill the missing values with the value that appears most often (aka "mode").

```python
df.isna().sum()
```

#### Yes there are missing values in this data in the 'ca' and 'thal' columns

```python
cols = ['ca', 'thal']

for col in cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

# Summarizing data

#### 1 What is the mean resting blood pressure for males and females?

```python
mean_bp= df['rest_bp'].mean()
mean_bp
```

```python
# Mean resting BP for males
male_mean = df[df['sex'] == 'Male']['rest_bp'].mean()

# Mean resting BP for females
female_mean = df[df['sex'] == 'Female']['rest_bp'].mean()

print("Mean resting BP for males:", male_mean)
print("Mean resting BP for females:", female_mean)
```

#### 2 What is the mean and median cholesterol levels for males and females?

```python
# Mean and median cholesterol for males
male_mean_chol = df[df['sex'] == 'Male']['chol'].mean()
male_median_chol = df[df['sex'] == 'Male']['chol'].median()

# Mean and median cholesterol for females
female_mean_chol = df[df['sex'] == 'Female']['chol'].mean()
female_median_chol = df[df['sex'] == 'Female']['chol'].median()

print("Male mean chol:", male_mean_chol)
print("Male median chol:", male_median_chol)
print("Female mean chol:", female_mean_chol)
print("Female median chol:", female_median_chol)
```

#### 3 Which age group has the largest median cholesterol levels for males?

```python
# Filter for males only
males = df[df['sex'] == 'Male']

# Group by age and calculate median cholesterol
median_chol_by_age = males.groupby('age')['chol'].median()

# Find the age (or age group) with the highest median cholesterol
highest_median_age = median_chol_by_age.idxmax()
highest_median_value = median_chol_by_age.max()

print("Age with highest median cholesterol (males):", highest_median_age)
print("Median cholesterol at that age:", highest_median_value)
```

#### 4 Compute mean risk value (the risk column was created in problem 2 of the "Manipulating data"section) for each age and sex. Which gender and age group has the highest average risk value? 

ùëéùëîùëí res_bp+ùëê‚Ñéùëúùëô+max_hr


```python
# Compute mean risk for each age and sex
risk_mean = df.groupby(['age', 'sex'])['risk'].mean().reset_index()

# Find the group with the highest average risk
highest_risk = risk_mean.loc[risk_mean['risk'].idxmax()]

print("Mean risk by age and sex:\n", risk_mean)
print("\nGroup with highest average risk:\n", highest_risk)
```
